```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
over = [x for x in np.arange(1,11,1)]
run_rate = [6,5,5,7,6.5,8,8.5,9,8,8.5]
run_rate2 = [5,4,5,6,6.5,7,8.5,9,8,8.5]
run_rate3 = [4,5,4,5,6.5,7,8.5,7,8.5,8]

font1 = {'family' : 'serif','color':'blue','size':20}
font2 = {'family' : 'serif','color':'darkred','size':15}

plt.title('Sports Data', fontdict=font1,loc='center')
plt.xlabel('Overs',fontdict=font2)
plt.ylabel('Run Rate',fontdict=font2)

plt.plot(over,run_rate)
plt.plot(over,run_rate2)
plt.plot(over,run_rate3)
plt.legend(['Run Rate','Run Rate 2','Run Rate 3'],loc='upper left')
plt.show()
```
