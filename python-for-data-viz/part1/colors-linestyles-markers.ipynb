```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# https://www.geeksforgeeks.org/python/linestyles-in-matplotlib-python/
```

```python
Wickets = [ x for x in range(1,11,1)]
scores = np.random.rand(10)*100
scorecard1 = scores ** 1.22
scorecard2 = scores ** 1.15
scorecard3 = scores * 2.5

font1={}
font2={}

plt.title('Title')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(Wickets,scorecard1,'b-o')
plt.plot(Wickets,scorecard2,color='c',linestyle='-.',marker='v')
plt.plot(Wickets,scorecard3,color='r',linestyle='--',marker='D')
plt.legend([1,2,3])
plt.show()
```
