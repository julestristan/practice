```python
import matplotlib.pyplot as plt
import numpy as np
import statistics
```

```python
languages = ['C','C++','Java','Python','PHP']
students = [23,17,35,29,12]
plt.bar(languages,students)
plt.show()
```

```python
height = [189,176,160,198,168,179,183,187,176,172,167,165,163,166,177,159,154,167]
plt.hist(height,bins=5,edgecolor='w',color='cyan')
m = statistics.mean(height)
sd = statistics.stdev(height)
plt.axvline(m,color = 'blue',linestyle='--')
plt.axvline(m+sd,color = 'black',linestyle='--')
plt.axvline(m-sd,color = 'black',linestyle='--')
plt.show()
```
