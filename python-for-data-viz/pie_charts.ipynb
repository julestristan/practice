```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.array([35,25,25,15])
mylabels = ['Apples','Bananas','Cherries','Dates']

plt.pie(y,labels=mylabels,wedgeprops={'linewidth':2, 'edgecolor':'white'},colors=['m','r','g','c'],autopct='%1.1f%%',startangle=0)
plt.legend()
plt.title('My chart')
plt.show()
```
